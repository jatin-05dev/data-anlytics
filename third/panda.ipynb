```python
import pandas as pd
import numpy as np

# DataFrame Creation
data = {
    "Name": ["Jatin", "Rahul", "Aman", "Riya"],
    "Age": [20, 21, 22, 19],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)

print("DataFrame:")
print(df)

# First 5 Rows
print("\nHead:")
print(df.head())

# Last 5 Rows
print("\nTail:")
print(df.tail())

# Shape
print("\nShape:")
print(df.shape)

# Columns
print("\nColumns:")
print(df.columns)

# Data Types
print("\nData Types:")
print(df.dtypes)

# Info
print("\nInfo:")
print(df.info())

# Statistics
print("\nDescribe:")
print(df.describe())

# Single Column
print("\nNames:")
print(df["Name"])

# Multiple Columns
print("\nName and Marks:")
print(df[["Name", "Marks"]])

# Row using loc
print("\nRow 0:")
print(df.loc[0])

# Row using iloc
print("\nSecond Row:")
print(df.iloc[1])

# Filter
print("\nMarks > 80:")
print(df[df["Marks"] > 80])

# Add New Column
df["Result"] = ["Pass", "Pass", "Pass", "Pass"]

print("\nAfter Adding Column:")
print(df)

# Mean
print("\nAverage Marks:")
print(df["Marks"].mean())

# Max Marks
print("\nMaximum Marks:")
print(df["Marks"].max())

# Min Marks
print("\nMinimum Marks:")
print(df["Marks"].min())

# Sort Values
print("\nSorted by Marks:")
print(df.sort_values("Marks"))

# Unique Values
print("\nUnique Ages:")
print(df["Age"].unique())

# Value Counts
print("\nAge Counts:")
print(df["Age"].value_counts())
```
