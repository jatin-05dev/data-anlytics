```python
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# 1. Line Plot
plt.figure(figsize=(5,3))
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()

# 2. Bar Chart
plt.figure(figsize=(5,3))
plt.bar(["A", "B", "C", "D"], [10, 20, 15, 25])
plt.title("Bar Chart")
plt.show()

# 3. Scatter Plot
plt.figure(figsize=(5,3))
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()

# 4. Histogram
data = [10,20,20,30,30,30,40,40,50]
plt.figure(figsize=(5,3))
plt.hist(data)
plt.title("Histogram")
plt.show()

# 5. Pie Chart
plt.figure(figsize=(5,3))
plt.pie(
    [40, 30, 20, 10],
    labels=["A", "B", "C", "D"],
    autopct="%1.1f%%"
)
plt.title("Pie Chart")
plt.show()
```
