```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print("Array:", arr)

# Basic Math
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Min:", np.min(arr))
print("Max:", np.max(arr))
print("Standard Deviation:", np.std(arr))
print("Variance:", np.var(arr))

# Power & Root
print("Square:", np.square(arr))
print("Cube:", np.power(arr, 3))
print("Square Root:", np.sqrt(arr))

# Logarithm
print("Log:", np.log(arr))
print("Log10:", np.log10(arr))

# Exponential
print("Exp:", np.exp(arr))

# Absolute Value
arr2 = np.array([-1, -2, 3, -4, 5])
print("Absolute:", np.abs(arr2))

# Trigonometry
angles = np.array([0, 30, 45, 60, 90])

print("Sin:", np.sin(np.radians(angles)))
print("Cos:", np.cos(np.radians(angles)))
print("Tan:", np.tan(np.radians(angles)))

# Rounding
nums = np.array([1.2, 2.7, 3.5, 4.9])

print("Round:", np.round(nums))
print("Floor:", np.floor(nums))
print("Ceil:", np.ceil(nums))

# Cumulative Operations
print("Cumulative Sum:", np.cumsum(arr))
print("Cumulative Product:", np.cumprod(arr))

# Difference
print("Difference:", np.diff(arr))
```

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

# Product
print("Product:", np.prod(arr))

# Index of Min & Max
print("Arg Min:", np.argmin(arr))
print("Arg Max:", np.argmax(arr))

# Percentile
print("50th Percentile:", np.percentile(arr, 50))

# Average
print("Average:", np.average(arr))

# Modulus
print("Mod:", np.mod(arr, 2))

# Remainder
print("Remainder:", np.remainder(arr, 2))

# GCD
print("GCD:", np.gcd(12, 18))

# LCM
print("LCM:", np.lcm(12, 18))

# Sign
a = np.array([-10, 0, 10])
print("Sign:", np.sign(a))

# Clip Values
print("Clip:", np.clip(arr, 2, 4))

# Truncate Decimal
b = np.array([1.9, 2.7, 3.5])
print("Trunc:", np.trunc(b))

# Degrees to Radians
print("Radians:", np.radians([0, 30, 45, 90]))

# Radians to Degrees
print("Degrees:", np.degrees(np.pi/2))

# Hyperbolic Functions
print("Sinh:", np.sinh(arr))
print("Cosh:", np.cosh(arr))
print("Tanh:", np.tanh(arr))
```

```python
import numpy as np

A = np.array([
    [1, 2],
    [3, 4]
])

# Matrix Transpose
print(A.T)

# Determinant
print(np.linalg.det(A))

# Inverse
print(np.linalg.inv(A))

# Eigen Values
print(np.linalg.eig(A))

# Matrix Rank
print(np.linalg.matrix_rank(A))

# Trace
print(np.trace(A))

# Matrix Multiplication
print(A @ A)
```
